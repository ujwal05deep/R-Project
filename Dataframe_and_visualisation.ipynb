```r
library(dplyr)
options(scipen=999)  # turn off scientific notation like 1e+06
library(ggplot2)     #package for Grammer of Graphics plot
```

```r
df1<-data.frame()  #create empty dataframe
```

```r
#user input for dataframe
n<-as.integer(readline("How many records do u want to add?"))
for(i in 1:n){
  {
     rollno<-as.integer(readline("Enter rollno"))
     name<-readline("Enter name")
     gender<-readline("Enter gender")
     sub1<-as.integer(readline("Enter markrs for subject1"))
     sub2<-as.integer(readline("Enter markrs for subject2"))
     sub3<-as.integer(readline("Enter markrs for subject3"))
     sub4<-as.integer(readline("Enter markrs for subject4"))
     sub5<-as.integer(readline("Enter markrs for subject5"))
  }
  total=sub1+sub2+sub3+sub4+sub5
  per=total/5
  cat("Hello",name,"you scored")
  cat("Total marks :",total)
  cat("Percentage :",per,"%")
  if(per>=75){
    grade='D'
    print(grade)
  }else if(per<=74 && per>=60){
    grade='A'
    print(grade)
  }else if(per<=59 && per>=50){
    grade='B'
    print(grade)
  }else if(per<=49 && per>=40){
    grade='C'
    print(grade)
  }else if(per<40){
    grade='F'
    print(grade)
  }
   #insert all user values into dataframe
  df<-data.frame(rollno,name,gender,sub1,sub2,sub3,sub4,sub5,total,per,grade)

  df1<-rbind(df1,df)  #joining two dataframes

}
```

```r
# Add label position for pie-chart
df1 <- df1 %>%
  arrange(desc(grade)) %>%
  mutate(lab.ypos = cumsum(per) - 0.5*per)
df1
```

```r
#pie chart displaying student grade according to percentage
ggplot(df1, aes(x = " ", y =per, fill =grade)) +
  geom_bar(width=1,stat = "identity", color = "black") +
  labs(title = "Student grade according to percentage") +
  geom_text(aes(y=lab.ypos,label = per), color = "black") +
  coord_polar(theta="y",start=0) +   #shape for pie chart
  theme_void()     #theme_void() to drop axes and labels
```
